## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY1="""
SELECT ConvertedComp FROM 
master"""

df_conv_comp= pd.read_sql_query(QUERY1,conn)
df_conv_comp.plot(kind="hist", bins=30, figsize=(5,5))
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY2="""
SELECT Age FROM 
master"""

df_age= pd.read_sql_query(QUERY2,conn)
df_age.plot(kind="box", figsize=(3,8))

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY3="""
SELECT Age, WorkWeekHrs FROM 
master"""

df_sp= pd.read_sql_query(QUERY3,conn)
df_sp.plot.scatter(x='Age',y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
QUERY4 = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
df_bp=pd.read_sql_query(QUERY4,conn)

sns.scatterplot(data=df_bp, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age', size=14) 
plt.xlabel('WorkWeekHrs', size=10) 
plt.ylabel('CodeRevHrs', size=10)

plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import matplotlib.pyplot as plt

QUERYx = """
SELECT 	DatabaseDesireNextYear, count(DatabaseDesireNextYear) as "Number of Respondents"
FROM DatabaseDesireNextYear
group by (DatabaseDesireNextYear)
order by "Number of Respondents" desc
limit 5
"""
df_piechart=pd.read_sql_query(QUERYx,conn)
label= df_piechart['DatabaseDesireNextYear']

plot= df_piechart.plot.pie(y='Number of Respondents', labels=label, autopct='%1.1f%%',radius=1.5, legend=None, ylabel=None)

plot.yaxis.set_label_coords(-0.4, 0.5)

plt.show()



```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERYxx = """
SELECT 	WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE age>=30 and age<=35
"""
df_schart=pd.read_sql_query(QUERYxx,conn)
df_schart1= df_schart.groupby('Age').median()
df_schart1.plot(kind='bar', stacked=True)

plt.show()



```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY_lc = """
SELECT 	ConvertedComp, Age
FROM master
WHERE age between 45 and 60
"""
df_lchart=pd.read_sql_query(QUERY_lc,conn)
df_lchart1= df_lchart.groupby('Age').median()
df_lchart1.plot(kind='line')

plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY_bc = """
SELECT 	MainBranch, count(MainBranch)
FROM master
group by MainBranch
"""
df_bchart=pd.read_sql_query(QUERY_bc,conn)
df_bchart.plot(kind='barh', x='MainBranch')
plt.show()

```
